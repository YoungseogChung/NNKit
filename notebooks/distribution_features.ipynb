## delta betan

```python
data_name = 'cnn_delta_full_pinj'
print('delta betan')

train_y = np.load(os.path.join(data_dir, data_name, 'train_y.npy'))
test_y = np.load(os.path.join(data_dir, data_name, 'test_y.npy'))

a = plt.hist(train_y, bins=100, label='train y', weights=np.ones(len(train_y)) / len(train_y), alpha=0.5)
b = plt.hist(test_y, bins=100, label='test y', weights=np.ones(len(test_y)) / len(test_y), alpha = 0.7)
plt.legend()
# plt.ylim(0, np.min([np.max(a[0]), np.max(b[0])]))
plt.xlim(np.max([np.min(a[1]), np.min(b[1])]), np.min([np.max(a[1]), np.max(b[1])]))
plt.xlabel('Data values')
plt.ylabel('Proportion of data')
plt.show()
print('Train target max:{0:8.3f},  min:{1:8.3f},  mean:{2:8.3f},  std:{3:8.3f}'.format(
    np.max(train_y), np.min(train_y), np.mean(train_y), np.std(train_y)))
print('Test target  max:{0:8.3f},  min:{1:8.3f},  mean:{2:8.3f},  std:{3:8.3f}'.format(
    np.max(test_y), np.min(test_y), np.mean(test_y), np.std(test_y)))

train_mean, train_std = np.mean(train_y), np.std(train_y)
train_norm = (train_y - train_mean)/train_std
test_norm = (test_y - train_mean)/train_std
a = plt.hist(train_norm, bins=100, label='train y', weights=np.ones(len(train_norm)) / len(train_norm), alpha=0.5)
b = plt.hist(test_norm, bins=100, label='test y', weights=np.ones(len(test_norm)) / len(test_norm), alpha = 0.7)
plt.legend()
# plt.ylim(0, np.min([np.max(a[0]), np.max(b[0])]))
plt.xlim(np.max([np.min(a[1]), np.min(b[1])]), np.min([np.max(a[1]), np.max(b[1])]))
plt.xlabel('Data values')
plt.ylabel('Proportion of data')
plt.show()

print('Train normalized target max:{0:8.3f},  min:{1:8.3f},  mean:{2:8.3f},  std:{3:8.3f}'.format(
    np.max(train_norm), np.min(train_norm), np.mean(train_norm), np.std(train_norm)))
print('Test normalized target  max:{0:8.3f},  min:{1:8.3f},  mean:{2:8.3f},  std:{3:8.3f}'.format(
    np.max(test_norm), np.min(test_norm), np.mean(test_norm), np.std(test_norm)))
```
